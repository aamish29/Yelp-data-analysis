# SQL Queries for business decision making

**Query 1: Analyze the current saturation of the Korean cuisine market. Here we would count the number of restaurants in Illinois with the category listed as “Korean”.**

```python
query = """
select city, count(*) from business
where state = 'IL' and categories like '%korean%'
group by city;
"""

df_korean_restaurants = pd.read_sql(query, engine)
print(df_korean_restaurants)
```

**Query 2 : Explore the Korean restaurants in O'Fallon, IL**

```python
query = """
select business_id, name from business
where state = 'IL' and city like '%Fallon' and categories like '%korea%';
"""

df_edwards_restaurants = pd.read_sql(query, engine)
print(df_edwards_restaurants)
```

So, there are 6 Korean restaurants in O'Fallon, IL, according to this data. Now, we would want to dig a bit deeper, by looking at these potential competitor's ratings, reviews, etc.

**Query 3 - Reviews of bb.q Chicken - O'Fallon**

```python
query = """
select * from review
where business_id='c-iLk7gTKR4naEgfvW4TzQ';
"""

df_bbqchicken_reviews = pd.read_sql(query, engine)
print(df_bbqchicken_reviews)
```

**Query 4 - Tips for bb.q Chicken - O'Fallon**

```python
query = """
select * from tip
where business_id='c-iLk7gTKR4naEgfvW4TzQ';
"""

df_bbqchicken_tips = pd.read_sql(query, engine)
print(df_bbqchicken_tips)
```

**Query 5 - Average Stars for bb.q Chicken - O'Fallon**

```python
query = """
select avg(stars) from review
where business_id='c-iLk7gTKR4naEgfvW4TzQ';
"""

df_bbqchicken_stars = pd.read_sql(query, engine)
print(df_bbqchicken_stars)
```

**Query 6 - Opening hours for the restaurants in O'Fallon**

```python
query = """
select hours from business
where state = 'IL' and city like '%fallon%' and categories like '%korea%';
"""

df_opening_hours = pd.read_sql(query, engine)
print(df_opening_hours)
```

So, it seems that the restaurants here don't open overnight, so there could be an opportunity there.

**Query 7 - Is there a demand of late night food in that city?**

```python
query = """
select text from review
where (text like '%late night%'
or text like '%overnight%'
or text like '%night%')

and business_id in (select business_id from business where state = 'IL' and city like '%fallon%');
"""

df_latenight_demand = pd.read_sql(query, engine)
print(df_latenight_demand)
```

So, it seems like there is some demand for overnight restaurants as most of the reviews here containing such keywords are negative. So, we can be profitable by opening an overnight Korean restaurant.

**Query 8 - Check for busiest postal codes in O'Fallon**

```python
query = """
select top 5 postal_code, count(business_id) as bus_cnt  from business where  city like '%fallon%'
group by postal_code order by bus_cnt desc;

"""

df_busy_codes = pd.read_sql(query, engine)
print(df_busy_codes)
```

**Query 9 - Check postal code of the restaurant we explored before.**
